```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

Se explora y transforma el archivo de Excel Interntet.xlsx, previamente a través de Excel se exploró las características de los datos y contenido para evitar transformar hojas que no se analizarán posteriormente en el archivo de EDA.

```python
# Leemos el archivo de excel Internet.xlx, hoja Velocidad_sin_Rangos.
df_Velocidad_sin_Rangos = pd.read_excel("../DataSet/Internet.xlsx", sheet_name="Velocidad_sin_Rangos")  
```

```python
# A continuación en cada celda usaremos funciones habituales para conocer la naturaleza de los datos y detectar posibles nulos, outliers, duplicados.
df_Velocidad_sin_Rangos.info()
```

```python
#Dejamos los valores nulos sin modificar porque en este caso representan un nro no muy alto y no serán analizadas las velocidades de este DF
df_Velocidad_sin_Rangos[df_Velocidad_sin_Rangos.isnull().any(axis=1)]
```

```python
df_Velocidad_sin_Rangos[df_Velocidad_sin_Rangos["Accesos"] <= 0]
```

```python
df_Velocidad_sin_Rangos = df_Velocidad_sin_Rangos[df_Velocidad_sin_Rangos["Accesos"] != 0]
```

```python
# Chequeamos valores únicos para evitar que se repitan con algún carácter diferente.

df_Velocidad_sin_Rangos['Provincia'].unique()
```

```python
# Encontramos que el DF tiene nombres de las provincias en mayúsculas en los registros del último año, no así en el resto de los registros,
# por lo que estandarizamos todo a maýuscula.

df_Velocidad_sin_Rangos["Provincia"] = df_Velocidad_sin_Rangos["Provincia"].str.upper()
```

```python
# Al encontrar que algunos registros tienen tilde en las provincias, se modifican tódos para que queden seteados y así evitar duplicidad,
# para esto se define un diccionario.

dicc_provincias = {"NEUQUÉN": "NEUQUEN", "CÓRDOBA": "CORDOBA", "CAPITAL FEDERAL": "CABA", 
                   "ENTRE RÍOS": "ENTRE RIOS", "RÍO NEGRO": "RIO NEGRO", "TUCUMÁN": "TUCUMAN"}

df_Velocidad_sin_Rangos['Provincia'] = df_Velocidad_sin_Rangos['Provincia'].replace(dicc_provincias)
```

```python
# Guardamos en csv
df_Velocidad_sin_Rangos.to_csv("../DataSet/Velocidad y Accesos por Provincias Historico.csv", encoding='utf-8', index=False)
```
